#1 - Análise do faturamento


```python
# Calcula o faturamento de cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento Loja 1: R${faturamento_loja1:,.2f}")
print(f"Faturamento Loja 2: R${faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R${faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R${faturamento_loja4:,.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
# Cria um gráfico de pizza com o faturamento de cada loja
faturamento_lojas = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(6, 6))  # Define o tamanho da figura
plt.pie(faturamento_lojas, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Faturamento por Loja')
plt.axis('equal')  # gráfico seja um círculo
plt.show()
```

# 2 - VENDAS POR CATEGORIA DE PRODUTO

```python
# Clcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

import pandas as pd

def analyze_products_by_category(df, loja_name):
    product_sales = df.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
    print(f"\nProdutos mais vendidos na {loja_name}:")
    print(product_sales)
    return product_sales

sales_loja1 = analyze_products_by_category(loja, "Loja 1")
sales_loja2 = analyze_products_by_category(loja2, "Loja 2")
sales_loja3 = analyze_products_by_category(loja3, "Loja 3")
sales_loja4 = analyze_products_by_category(loja4, "Loja 4")
```

```python
# Criar um gráfico de barras verticais que agrupe os produtos mais vendidos nas lojas

import matplotlib.pyplot as plt

# Criar o gráfico de barras verticais
fig, ax = plt.subplots(figsize=(10, 6))

width = 0.15
x = range(len(sales_loja1.index))

rects1 = ax.bar([i - 1.5 * width for i in x], sales_loja1.values, width, label='Loja 1')
rects2 = ax.bar([i - 0.5 * width for i in x], sales_loja2.values, width, label='Loja 2')
rects3 = ax.bar([i + 0.5 * width for i in x], sales_loja3.values, width, label='Loja 3')
rects4 = ax.bar([i + 1.5 * width for i in x], sales_loja4.values, width, label='Loja 4')


ax.set_ylabel('Quantidade Vendida')
ax.set_title('Vendas por Categorias por Loja')
ax.set_xticks(x)
ax.set_xticklabels(sales_loja1.index, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

# 4 - PRODUTOS MAIS E MENOS VENDIDOS POR LOJA

```python
def analyze_top_bottom_products(df, loja_name):
    # Conta as ocorrências de cada produto para determinar a quantidade vendida
    product_sales = df.groupby('Produto')['Produto'].count()

    # Ordena as vendas em ordem decrescente para encontrar os mais vendidos
    top_products = product_sales.sort_values(ascending=False).head(5)  # Pega os 5 mais vendidos

    # Ordena as vendas em ordem crescente para encontrar os menos vendidos
    bottom_products = product_sales.sort_values(ascending=True).head(5)  # Pega os 5 menos vendidos

    print(f"\nProdutos mais vendidos na {loja_name}:")
    print(top_products)

    print(f"\nProdutos menos vendidos na {loja_name}:")
    print(bottom_products)

    return top_products, bottom_products


top_products_loja1, bottom_products_loja1 = analyze_top_bottom_products(loja, "Loja 1")
top_products_loja2, bottom_products_loja2 = analyze_top_bottom_products(loja2, "Loja 2")
top_products_loja3, bottom_products_loja3 = analyze_top_bottom_products(loja3, "Loja 3")
top_products_loja4, bottom_products_loja4 = analyze_top_bottom_products(loja4, "Loja 4")



```

```python
# Gráficos que  sintetizam os produtos mais vendidos por loja

import matplotlib.pyplot as plt

def plot_top_products_per_store(sales_data_list, store_names):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
    axes = axes.flatten()  # Achata a matriz de eixos para facilitar a iteração

    for i, (sales_data, store_name) in enumerate(zip(sales_data_list, store_names)):
        ax = axes[i]
        sales_data.plot(kind='bar', ax=ax)
        ax.set_title(f'Produtos Mais Vendidos - {store_name}')
        ax.set_xlabel('Produto')
        ax.set_ylabel('Quantidade Vendida')
        ax.tick_params(axis='x', rotation=45) # Rotaciona os rótulos do eixo x para melhor visualização


    plt.tight_layout()
    plt.show()

# Exemplo de uso (substitua pelos seus dados reais):
store_names = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
plot_top_products_per_store([top_products_loja1, top_products_loja2, top_products_loja3, top_products_loja4], store_names)
```

```python
# Gráfico que sintetizam produtos menos vendidos por lojas

import matplotlib.pyplot as plt

def plot_bottom_products_per_store(sales_data_list, store_names):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
    axes = axes.flatten()  # Achata a matriz de eixos para facilitar a iteração

    for i, (sales_data, store_name) in enumerate(zip(sales_data_list, store_names)):
        ax = axes[i]
        sales_data.plot(kind='bar', ax=ax, color='skyblue') # Define a cor das barras
        ax.set_title(f'Produtos Menos Vendidos - {store_name}', fontsize=10) # Ajusta o tamanho do título
        ax.set_xlabel('Produto', fontsize=8) # Ajusta o tamanho do rótulo do eixo x
        ax.set_ylabel('Quantidade Vendida', fontsize=8) # Ajusta o tamanho do rótulo do eixo y
        ax.tick_params(axis='x', rotation=45, labelsize=7) # Ajusta o tamanho e rotação dos rótulos do eixo x
        ax.tick_params(axis='y', labelsize=7) # Ajusta o tamanho dos rótulos do eixo y
        ax.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona uma grade no eixo y


    plt.tight_layout()
    plt.show()

# Exemplo de uso (substitua pelos seus dados reais):
store_names = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
plot_bottom_products_per_store([bottom_products_loja1, bottom_products_loja2, bottom_products_loja3, bottom_products_loja4], store_names)
```
